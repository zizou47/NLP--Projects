```python
import pandas as pd
import numpy as np
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
 import tensorflow as tf
print(tf.__version__)
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
### Vocabulary size
voc_size=500
```

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences  # this make vector same size by adding zeros in beginning or in last
from tensorflow.keras.models import Sequential
```

```python
## pre padding result
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## 10 feature dimesnions like word2vec and here each number likr (0) with 10 diffrent representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
##'the glass of milk',
embedded_docs[0]
```

```python
import tensorflow as tf

# Ensure embedded_docs is a tf.Tensor
# The Embedding layer expects integer inputs, so we specify dtype=tf.int32
tf_embedded_docs = tf.convert_to_tensor(embedded_docs, dtype=tf.int32)
print(model.predict(tf_embedded_docs))
```
