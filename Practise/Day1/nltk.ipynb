```python
paragraph = """"
Stemming is well-suited for scenarios where speed and scalability are prioritized over precise accuracy. It is particularly effective in search systems and text retrieval applications where an approximate matching of words is sufficient. By reducing words to their root forms, stemming enhances the efficiency of these systems while managing large volumes of text."""
```

```python
paragraph
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
## tokenization : convert paragraph > sentences> words
nltk.download('punkt')
sentences = nltk.sent_tokenize(paragraph)
```

```python
print(sentences)
```

```python
stemmer = PorterStemmer()
```

```python
stemmer.stem('emailing')
```

```python
#lemmatazation
from nltk.stem import WordNetLemmatizer
lemmitazer = WordNetLemmatizer()
```

```python
lemmitazer.lemmatize('goes')
```

```python
# cleaning ur txt
import re
corpus= []
for i in range (len(sentences)):
    text = re.sub('[^a-zA-Z]', ' ', sentences[i]) # if u found rather then letter replace with space 
    text = text.lower()
    corpus.append(text)
```

```python
corpus
```

```python
from nltk.tokenize import word_tokenize
```

```python
# stemming
stop_words = set(stopwords.words('english'))
for sentence in corpus:
    # tokenise sentences int word
    words = word_tokenize(sentence)
    stemming = [stemmer.stem(word) for word in words if word.lower() not in stop_words]
    print (stemming)
```

```python
# lemmatazation

stop_words = set(stopwords.words('english'))
for sentence in corpus:
    # tokenise sentences int word
    words = word_tokenize(sentence)
    lemmatazation = [lemmitazer.lemmatize(word) for word in words if word.lower() not in stop_words]
    print (lemmatazation)
```
